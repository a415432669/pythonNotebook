```python
import requests
from pyquery import PyQuery as pq
```

```python
res = requests.get('http://www.ygdy8.net/html/gndy/oumei/list_7_1.html')
res.encoding = 'gb2312'
doc = pq(res.text)
arrList = doc('.co_content8 tr td b a:nth-child(2)').attr('href')
arrList
```

```python
def getPage(pageNum):
    res = requests.get('http://www.ygdy8.net/html/gndy/oumei/list_7_%s.html'%pageNum)
    res.encoding = 'gb2312'
    doc = pq(res.text)
    arrList = doc('.co_content8 tr td b a:nth-child(2)')
    for item in arrList.items():
        #print('http://www.ygdy8.net'+item.attr('href'))
        yield 'http://www.ygdy8.net'+item.attr('href')
#getPage(1)
```

```python
def saveMysql(result):
    db = pymysql.connect(host='127.0.0.1',user='root',passwd ='123456',db='films',charset='utf8')
    cursor = db.cursor()
    sql = "insert into app_filminfo (title,datatime,brief,content) values ('%s','%s','%s','%s')"%(result[0][0],result[0][1],result[0][2],result[0][3])
    print(sql)
    try:
        cursor.execute(sql)
        db.commit()
    except Exception as err:
        print('出错了')
        print(err)
    db.close()
```

```python
def parsePage(url):
    res = requests.get(url)
    res.encoding = 'gb2312'
    pattern = re.compile('<div.*?title_all.*?font.*?>(.*?)</font>.*?发布时间：(.*?)\s.*?简.*?介.*?<br /><br />\s*(.*?)<br />.*?bgcolor="#fdfddf"><a href="(.*?)"',re.S)
    result = re.findall(pattern,res.text)
    return result
```

```python
def getPageNum(url):
    res = requests.get(url)
    res.encoding = 'gb2312'
    pattern = re.compile('<div class="x".*?bgcolor="#F4FAE2">.*?共(.*?)页',re.S)
    num = re.search(pattern,res.text).group(1)
    return int(num)
getPageNum('http://www.ygdy8.net/html/gndy/oumei/index.html')
```

```python
import requests
from pyquery import PyQuery as pq
import re

pageNum = getPageNum('http://www.ygdy8.net/html/gndy/oumei/index.html')
for index in range(1,pageNum+1):
    for item in getPage(index):
        result = parsePage(item)
        saveMysql(result)
        print('保存成功：',result[0][0])
```
