```python
#要使用必先导入
import pygame
#导入pygame中的所有常量
from pygame.locals import *
import sys


#初始化PyGame
pygame.init()
#创建一个600*500的窗口
screen = pygame.display.set_mode((600,500))

#打印文字到窗口
myfont = pygame.font.Font(None,60)
#RGB,R红色，G绿色，B蓝色
white = (255,255,255)
blue = (0,0,255)
textImage = myfont.render('Hello PyGame',True,white)

while True:
    for event in pygame.event.get():
        if event.type in (QUIT,KEYDOWN):
            sys.exit()
    # 清理屏幕，绘制，然后刷新
    screen.fill(blue)
    screen.blit(textImage, (100, 100))
    pygame.display.update()
```

```python
#要使用必先导入
import pygame
#导入pygame中的所有常量
from pygame.locals import *
import sys
import random

#初始化PyGame
pygame.init()
#创建一个600*500的窗口
screen = pygame.display.set_mode((600,500))
# help(pygame.display.set_mode)

# myfont1 = pygame.font.Font(None,50)
blue = (255,255,255)
# fontShow = myfont1.render('helloWorld',True,blue)

black = (0,0,0)
red = (100,0,0)
purple = (100,0,100)
pos = 200,200,150,50
n= 1
count = 5
cNum = 0
clock = pygame.time.Clock()
L = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
randonAlpha = L[random.randint(0,25)]
print(randonAlpha)
myfont = pygame.font.Font(None,60)
#RGB,R红色，G绿色，B蓝色
white = (255,255,255)
textImage = myfont.render(randonAlpha,True,white)
wordCount = 0
isShow = 1

while True:
#     print(random.randint(1,25))
    
    for event in pygame.event.get():
        if event.type in (QUIT,):
            sys.exit()
            
        if event.type in (MOUSEMOTION,):
            circlepos = event.pos
            
        if event.type in (KEYDOWN,):
            print(event.unicode)
            if event.unicode == randonAlpha:
                print('字母成功销毁...')
                myfont = pygame.font.Font(None,60)
                #RGB,R红色，G绿色，B蓝色
                white = (255,255,255)
                textImage = myfont.render(randonAlpha,True,white)
                wordCount = wordCount+1
                randonAlpha = L[random.randint(0,25)]
                print('新的字母以生成：%s'%randonAlpha)
                
    if pygame.time.get_ticks()/1000>10 and isShow:
        print('每分钟打字速度为：%s'%(wordCount*6))
        isShow = 0
                
                
            
    screen.fill(black)
    screen.blit(textImage,(100,100))
    pos = n,200,150,50
    color = n,100,n
    
    if cNum == count:
#         print(clock.tick())
        n = n + 1
        cNum=0
    cNum = cNum + 1
    if n>255:
        n=1
    pygame.draw.rect(screen,red,pos,0)
    pygame.draw.circle(screen,color,circlepos, 50, 0)
    pygame.display.update()
    

    
```

```python
#要使用必先导入
import pygame
#导入pygame中的所有常量
from pygame.locals import *
import sys


#初始化PyGame
pygame.init()
#创建一个600*500的窗口
screen = pygame.display.set_mode((600,500))

#打印文字到窗口
myfont = pygame.font.Font(None,60)
#RGB,R红色，G绿色，B蓝色
white = (255,255,255)
blue = (0,0,255)
textImage = myfont.render('Hello PyGame',True,white)
spaceImg = pygame.image.load("./test.png").convert_alpha()
# spaceImg = pygame.transform.smoothscale(logo,(200,200))
spaceChile = spaceImg.subsurface((0,0,100,100))
tt = 0
posChild = 0

while True:
    for event in pygame.event.get():
        if event.type in (QUIT,KEYDOWN):
            sys.exit()
    # 清理屏幕，绘制，然后刷新
    currentTime = pygame.time.get_ticks()
    if int(currentTime/125)> tt:
        tt = int(currentTime/125)
        spaceChile = spaceImg.subsurface((posChild,0,100,100))
        posChild= posChild+100
        if posChild>300:
            posChild = 0
        
    
    screen.fill(blue)
    screen.blit(textImage, (100, 100))
    screen.blit(spaceChile,(200,200))
    pygame.display.update()
```
