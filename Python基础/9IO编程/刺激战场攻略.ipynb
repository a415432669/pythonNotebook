````python
import re
import json
````

````python
pattern = re.compile('<li>.*?href="(/gicp/news/104/6.*?)".*?gllink.*?>\s*(.*?)\s*</a>.*?<span>\s*(.*?)\s*</span>',re.S)
result = re.findall(pattern,res.text)
result = json.dumps(result)


with open('ciji.json','w') as f:
    f.write(result)
````
