````python
#要使用必先导入
import pygame
#导入pygame中的所有常量
from pygame.locals import *
import sys


#初始化PyGame
pygame.init()
#创建一个600*500的窗口
screen = pygame.display.set_mode((600,500))

#打印文字到窗口
myfont = pygame.font.Font(None,60)
#RGB,R红色，G绿色，B蓝色
white = (255,255,255)
blue = (0,0,255)
textImage = myfont.render('Hello PyGame',True,white)

while True:
    for event in pygame.event.get():
        if event.type in (QUIT,KEYDOWN):
            sys.exit()
    # 清理屏幕，绘制，然后刷新
    screen.fill(blue)
    screen.blit(textImage, (100, 100))
    pygame.display.update()
````

````python
#main.py
#要使用必先导入
import pygame
#导入pygame中的所有常量
from pygame.locals import *
import sys



            
            
    
# class Scene():
#     def __init__(self,name):
#         self.sceneName = name
#         pass
#     def event(self,eventList):
#         pass
#     def show(self):
#         pass
    

class startScene():
    def __init__(self,name):
        self.sceneName = name
        self.back = pygame.image.load("./image/end.jpg").convert_alpha()
        pass
    def event(self,eventList,app):
        for event in eventList:
            if event.type in (QUIT,):
                sys.exit()
            if event.type in (KEYDOWN,):
                app.sceneIndex = 1
                
        pass
    def show(self,app):
        #打印文字到窗口
        myfont = pygame.font.Font(None,60)
        #RGB,R红色，G绿色，B蓝色
        white = (255,255,255)
        blue = (0,0,255)
        
        backimage = pygame.transform.scale(self.back, (800, 500))
        textImage = myfont.render('Start Game',True,white)
        app.screen.fill(blue)
        app.screen.blit(backimage, (0, 0))
        app.screen.blit(textImage, (100, 100))
        pygame.display.update()
        
        pass

class GameingScene():
    def __init__(self,name):
        self.sceneName = name
        pass
    def event(self,eventList,app):
        for event in eventList:
            if event.type in (QUIT,):
                sys.exit()
            if event.type in (KEYDOWN,):
                if event.unicode == 'k': 
                    app.sceneIndex = 2
                
        pass
    def show(self,app):
        #打印文字到窗口
        myfont = pygame.font.Font(None,60)
        #RGB,R红色，G绿色，B蓝色
        white = (255,255,255)
        blue = (0,0,255)
        textImage = myfont.render('Gaming Scene',True,white)
        app.screen.fill(blue)
        app.screen.blit(textImage, (100, 100))
        pygame.display.update()
        
        pass

class endGameScene():
    def __init__(self,name):
        self.sceneName = name
        pass
    def event(self,eventList,app):
        
        for event in eventList:
            if event.type in (QUIT,):
                sys.exit()
            if event.type in (KEYDOWN,):
                if event.unicode == '1':
                    app.sceneIndex = 0
                        
                
        pass
    def show(self,app):
#         timer = timeClock()
        #打印文字到窗口
        myfont = pygame.font.Font(None,60)
        #RGB,R红色，G绿色，B蓝色
        white = (255,255,255)
        blue = (0,0,255)
        textImage = myfont.render('end Game',True,white)
        app.screen.fill(blue)
        if timeClock.timing(1000):
            print('每隔一秒后执行')
        app.screen.blit(textImage, (100, 100))
        pygame.display.update()
        
        pass
    
    
class timeClock():
    startTime = 0
    endTime = 0
    state = True
    def __init__(self):
        pass
    def timing(sceonds):
#         print(timeClock.state)
        if timeClock.state == True:
            timeClock.startTime = pygame.time.get_ticks()
            timeClock.state = False
#             print(self.state)
            return True
        else:
#             print(timeClock.endTime)
#             print(timeClock.startTime)
            timeClock.endTime = pygame.time.get_ticks()
            timeClock.costTime =timeClock.endTime - timeClock.startTime
            if timeClock.costTime<sceonds:
                timeClock.state = False
                return False
            else:
                timeClock.startTime = 0
                timeClock.state = True
                return False
        

        
        
class System():
    def __init__(self):
        self.sceneIndex = 0
#         startGame = Scene('开始界面')
#         gameing = Scene('游戏中')
#         endGame = Scene('游戏结束界面')
        
        
        pygame.init()
        #创建一个600*500的窗口
        self.screen = pygame.display.set_mode((600,500))
        startGame = startScene('开始界面')
        gameing = GameingScene('游戏中')
        endGame = endGameScene('游戏结束界面')
        self.sceneList = [startGame,gameing,endGame]
        pass
        
    def run(self):
        #初始化PyGame
        i = 0
        while True:
            #场景调用
            i = i + 1
            print(i)
            app = self
            eventList = pygame.event.get()
            sceneNow = self.sceneList[self.sceneIndex]
            sceneNow.event(eventList,app)
            sceneNow.show(app)

        
app = System()
app.run()
    
````
